# 컴퓨터 비전이란?
- 이미지나 영상과 같은 시각 데이터를 컴퓨터가 입력받아 이해하도록 처리하는 분야.
- 객체를 식별하고 분류하는 방법으르 학습
- 사람또는 사물을 감지하거나 인식하며 움직임까지 추적할 수 있다.


### < 컴퓨터 비전 종류 >

image.png

- **분류(Classification)**는 해당 이미지가 어떤 객체(물체)를 나타내는지 분류
- **Localization**은 이미지 안에 있는 하나의 객체 위치를 찾아주는 작업. 객체 위치를 표시하는 네모 박스를 ** 경계 박스(Bounding Box)**라고 합니다.
- **객체 탐지(Object Detection)**는 이미지 안에 있는 여러 객체들의 위치를 찾아내고, 각각 분류까지 하는 작업
- **세그멘테이션(Segmentation)**은 객체 탐지보다 더 발전된 단계로, 픽셀 단위로 객체를 탐지하는 작업

### *Q) 컴퓨터 비전 활용 분야는 무엇이 있을까?*

**자율주행 자동차**
컴퓨터 비전은 자율주행차가 탐색하고 실시간 결정을 내리는 데 중추적인 역할을 합니다. 카메라, LiDAR 및 센서를 사용하여 차량은 물체, 보행자 및 교통 표지판을 식별하여 안전하고 효율적인 여행을 보장할 수 있습니다.

**의료 영상**
의료 분야에서 컴퓨터 비전은 X-레이 및 MRI와 같은 의료 이미지를 분석하여 질병 진단을 지원합니다. AI 기반 알고리즘은 이상 징후를 감지하여 의료 전문가가 정확하고 시기적절한 진단을 내릴 수 있도록 지원합니다.

**영상 복원 및 강화**
영상 복원은 손상된 또는 노이즈가 있는 영상을 복구하는 기술, 그리고 영상 강화는 낮은 해상도의 영상을 고해상도로 변환하는 기술입니다. 이러한 기술은 의료 영상, 보안 영상, 영화 산업 등 다양한 분야에서 활용됩니다.

**영상 분류 및 분할**
영상 분류는 영상 속에서 특정 객체를 식별하고 분류하는 작업으로, 객체 인식과 함께 광고 산업, 상품 추천 등 다양한 분야에서 적용됩니다.
영상 분할은 영상을 여러 영역으로 구분하는 기술로, 의료 영상, 산업 영상 검사 등 다양한 분야에서 활용됩니다.

**스마트 상품 추천**
컴퓨터 비전을 사용한 스마트 상품 추천은 온라인 쇼핑에서 고객의 선호도를 분석하여 맞춤형 추천 상품을 제공하는데 큰 도움을 줍니다. 상품 이미지를 기반으로 유사한 상품을 찾거나, 고객이 원하는 핵심 요소를 파악해 상품을 추천할 수 있습니다.

**얼굴 인식 및 생체 인증**
얼굴 인식의 경우 카메라와 컴퓨터 비전을 이용, 개인의 얼굴 학습 및 인식이 가능하며, 이 기능을 활용한 기기 해금, 장치 허가 등 다양한 분야에서 사용됩니다.

**공업 자동화**
컴퓨터 비전은 결함이 있는 제품을 검사하고, 로봇 팔을 안내하고, 일관된 품질 관리를 보장함으로써 제조 공정을 최적화합니다. 그 결과 효율성이 증가하고 인적 오류가 감소합니다.

**보안 및 검사**
컴퓨터 비전으로 구동되는 감시 시스템은 의심스러운 활동에 대해 환경을 모니터링합니다. 비디오 피드를 분석함으로써 AI 알고리즘은 잠재적인 위협을 식별하고 보안 담당자에게 경고할 수 있습니다.

**농업**
컴퓨터 비전은 작물의 건강과 성장을 평가하여 정밀 농업을 지원합니다. 카메라가 장착된 드론은 이미지를 캡처할 수 있으므로 농부는 주의가 필요한 영역을 식별하고 작물 수확량을 최적화할 수 있습니다.

**증강 현실(AR) 및 가상 현실(VR)**
AR 및 VR 경험은 디지털 요소를 현실 세계와 혼합하기 위해 컴퓨터 비전에 크게 의존합니다. 카메라는 움직임을 추적하여 사용자가 가상 ​​개체 및 환경과 원활하게 상호 작용할 수 있도록 합니다.

**미디어 산업 콘텐츠 생산**
컴퓨터 비전을 활용한 프레이밍 및 콘텐츠 생산은 영화, 광고, 게임 산업에서 다양한 형태로 활용됩니다. 요즘 실시간 그래픽 생성, 적대적 생성 네트워크(GAN) 기반의 이미지 생성 등 다양한 콘텐츠 프로덕션에 컴퓨터 비전 기술이 사용되고 있습니다.

**광학 문자 인식**
광학 문자 인식(OCR)은 컴퓨터 비전 기술의 한 분야로, 텍스트 이미지를 컴퓨터가 읽을 수 있는 텍스트 포맷으로 변환하는 기술입니다

# 컴퓨터에서 이미지를 표현하는 방법
컴퓨터에서 표현하는 이미지 데이터는 픽셀(PIXEL)의 모음이다. 컴퓨터에서는 이미지를 숫자로 받아들이며, 이 숫자를 이용해 연산을 수행할 수 있다.
- **흑백사진의 저장 방법:**흑백 사진의 경우 흰색과 검정색 사이의 여러 가지 음영의 회색이 존재하며, 0(검은색)~255(흰색)까지의 숫자로 표현될 수 있다.

image.png

- **컬러사진의 저장 방법:** 색상이 있는 사진은 빨강(R), 초록(G), 파랑(B)이 혼합되어 색을 표현한다. 3가지 색상이 0~255정수로 표현된다. 한 픽셀당 3가지 색을 표현하도록 3개의 숫자가 필요하다.

필터로 사진에 다양한 효과 주기.

```python
# 기본 행렬
import numpy as np

# 이미지 강렬하게 만드는 필터
a = np.array( [[-1,-1,-1],
             [0,5,0],
             [1,1,1]] )

# 이미지 블러처리 하게 만드는 필터
#a =  np.ones((5,5), np.float32)/ 25

# 이미지에 만든 필터 씌우기
blured = cv2.filter2D(img, -1, a)

# 필터 씌운 이미지 띄우기
cv2_imshow(blured)
```
